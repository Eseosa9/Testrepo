## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

table_name = 'ConvertedComp'


QUERY = """
SELECT {} as ConvertedComp
FROM master
""".format(table_name)

Histogram_df = pd.read_sql_query(QUERY,conn)
Histogram_df.hist(bins=15)
plt.show()

```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

table_name = 'Age'


QUERY = """
SELECT {} as Age
FROM master
""".format(table_name)

Box_df = pd.read_sql_query(QUERY,conn)
Box_df.boxplot(color = 'Blue')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
Scatter_df=pd.read_sql_query(QUERY,conn)
Scatter_df
Scatter_df.plot(kind='scatter', x='Age', y='WorkWeekHrs',s = 200, figsize=(10, 8), alpha=0.5, color='red')

Scatter_df.corr()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""
Bubble_df=pd.read_sql_query(QUERY,conn)
Bubble_df
Bubble_df.plot(kind='scatter', x='WorkWeekHrs',y='Age',s = 200, figsize=(10, 8), alpha=0.5, color='red')
Bubble_df.plot(kind='scatter', x='CodeRevHrs',y='Age',s = 200, figsize=(10, 8), alpha=0.5, color='green')
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """
select count(DatabaseDesireNextYear) as Count, DatabaseDesireNextYear from DatabaseDesireNextYear group by DatabaseDesireNextYear order by count(DatabaseDesireNextYear) DESC LIMIT 5
"""

df2=pd.read_sql_query(QUERY,conn)
df2.set_index('DatabaseDesireNextYear',inplace=True)
df2

labels = df2['Count']
sizes = df2['Count']

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
print(df2)


QUERY_B = """
select count(LanguageDesireNextYear) as Count, LanguageDesireNextYear from LanguageDesireNextYear group by LanguageDesireNextYear order by count(LanguageDesireNextYear) DESC LIMIT 5
"""
df3=pd.read_sql_query(QUERY_B,conn)
df3.set_index('LanguageDesireNextYear',inplace=True)
df3

labels_B = df3['Count']
sizes_B = df3['Count']

fig2, ax2 = plt.subplots()
ax2.pie(sizes_B, labels=labels_B, autopct='%1.1f%%',shadow=True, startangle=90)
ax2.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
print(df3)

QUERY_C = """
select count(LanguageWorkedWith) as Count, LanguageWorkedWith from LanguageWorkedWith group by LanguageWorkedWith order by count(LanguageWorkedWith) DESC LIMIT 5
"""
df4=pd.read_sql_query(QUERY_C,conn)
df4.set_index('LanguageWorkedWith',inplace=True)
df4

labels_C = df4['Count']
sizes_C = df4['Count']

fig3, ax3 = plt.subplots()
ax3.pie(sizes_C, labels=labels_C, autopct='%1.1f%%',shadow=True, startangle=90)
ax3.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
print(df4)





QUERY_D = """
select count(DatabaseWorkedWith) as Count, DatabaseWorkedWith from DatabaseWorkedWith group by DatabaseWorkedWith order by count(DatabaseWorkedWith) DESC LIMIT 5
"""
df5=pd.read_sql_query(QUERY_D,conn)
df5.set_index('DatabaseWorkedWith',inplace=True)
df5

labels_D = df5['Count']
sizes_D = df5['Count']

fig4, ax4 = plt.subplots()
ax4.pie(sizes_D, labels=labels_D, autopct='%1.1f%%',shadow=True, startangle=90)
ax4.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
print(df5)

QUERY_E = """
select count(MiscTechWorkedWith) as Count, MiscTechWorkedWith from MiscTechWorkedWith group by MiscTechWorkedWith order by count(MiscTechWorkedWith) DESC LIMIT 5
"""
df6=pd.read_sql_query(QUERY_E,conn)
df6.set_index('MiscTechWorkedWith',inplace=True)
df6

labels_E = df6['Count']
sizes_E = df6['Count']

fig5, ax5 = plt.subplots()
ax5.pie(sizes_E, labels=labels_E, autopct='%1.1f%%',shadow=True, startangle=90)
ax5.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
print(df6)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master WHERE Age BETWEEN 30 and 35
"""

df3 = pd.read_sql_query(QUERY, conn)
df3

df_median = df3.groupby(df3.Age)[['WorkWeekHrs','CodeRevHrs']].median()
df_median

df_median.plot(kind='bar')
plt.title('Median of WorkWeekHrs and CodeRevHrs for the age of 30 and 35')
plt.show()


```
